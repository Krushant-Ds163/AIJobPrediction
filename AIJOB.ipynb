```python
df.head()
```

```python
industry_salary=df.groupby('Industry')['Salary_USD'].mean()
top_10_industry=industry_salary.sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
top_10_industry.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Top 10 Highest Paying AI Jobs by Industry', fontsize=16)
plt.xlabel('Industry', fontsize=12)
plt.ylabel('Average Salary in USD', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



```

```python
heatmap_data=pd.crosstab(df['Required_Skills'],df['AI_Adoption_Level'])
plt.figure(figsize=(10,6))
sns.heatmap(heatmap_data, cmap='Blues',annot=True, fmt='d', linewidths=0.5)
plt.title('Skills vs AI Adoption Level', fontsize=16)
plt.xlabel('AI Adoption Level', fontsize=12)
plt.ylabel('Required Skills', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
location_Salary=df.groupby('Location')['Salary_USD'].mean()
top_10_location=location_Salary.sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
top_10_location.plot(kind='bar', edgecolor='black')
plt.title("Top 10 Location Highest Paying Salary")
plt.xlabel("Location")
plt.ylabel("Salary In USD")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
remote_friendly_data = df[df['Remote_Friendly'] == 'Yes']
industry_counts = remote_friendly_data['Industry'].value_counts()
plt.figure(figsize=(8, 8))
plt.title("Percentage of remote-friendly AI jobs across industries")
plt.pie(industry_counts, labels=industry_counts.index, autopct='%1.1f%%')
plt.tight_layout()
plt.show()
```

```python
sorted_data=df.sort_values(by='Job_Growth_Projection',ascending=False)
plt.figure(figsize=(10,6))
plt.bar(sorted_data['Job_Title'],sorted_data['Job_Growth_Projection'])
plt.title('Job Titles vs. Job Growth Projection', fontsize=16)
plt.xlabel('Job Titles', fontsize=12)
plt.ylabel('Job Growth Projection (%)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
grouped_data = df.groupby(['Location', 'Industry'])['Salary_USD'].mean()
heatmap_data = grouped_data.unstack()
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, linewidths=0.5)
plt.title('Highest Paying Industry by Location')
plt.xlabel('Industry')
plt.ylabel('Location')
plt.tight_layout()
plt.show()
```

```python
avg_salary=df.groupby('Required_Skills')['Salary_USD'].mean().sort_values(ascending=False)
#avg_salary_sorted=avg_salary.sort_values(by='Salary_USD', ascending=False)
plt.figure(figsize=(10,6))
avg_salary.plot(kind='bar', edgecolor='black')
plt.title('Required Skills vs. Average Salary')
plt.xlabel('Required Skills')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
